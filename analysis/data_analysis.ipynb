```python
import matplotlib.pyplot as plt
import seaborn as sns

singleham = [(1, 4.9303e+01),
             (2, 9.4488e+01),
             (4, 1.8606e+02),
             (8, 3.4283e+02),
             (16, 6.7301e+02),
             (32, 1.1861e+03),
             (54, 1.6904e+03),
             (64, 1.7945e+03),
             (128, 2.2294e+03)]


gcp_box = [(54 ,8.4089e+02),
(32 ,5.2515e+02),
(16 ,2.7361e+02),
(8  ,1.4117e+02),
(4  ,7.1235e+01),
(2  ,3.6741e+01),
(1  ,1.8667e+01),]

gcp_box.reverse()


1.6904e+03/4.9303e+01

# Calculate the normalized and efficiency values
for value in singleham: 
    print(f"{value[0]}: {value[1]/singleham[0][1]}, {(value[1]/singleham[0][1])/value[0]}")

x = [value[0] for value in singleham]
y = [value[1]/singleham[0][1] for value in singleham]
ideal = [value[0] for value in singleham]

ygcp = [value[1]/gcp_box[0][1] for value in gcp_box]
xgcp = [value[0] for value in gcp_box]


# Create the plot
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.lineplot(x=x, y=y, label='ham_actual', marker='D')

ax = sns.lineplot(x=x, y=ideal, label='Ideal', linestyle='--')
ax = sns.lineplot(x=xgcp, y=ygcp, label='gcp_actual', marker='D')

ax.set_xlabel('Number of nodes')
ax.set_ylabel('Time (log scale)')
ax.set_yscale('log', base=2)
ax.set_xscale('log', base=2)

ax.axvline(x=54, color='r', linestyle='dashed')
ax.text(54.5, 100, '54 Cores', rotation=0)


# ax.set_xticklabels([1,2,4,8,16,32,64])
# ax.set_yticklabels([1,2,4,8,16,32,64,128])

plt.xticks(x)
plt.legend()
plt.show()
```
